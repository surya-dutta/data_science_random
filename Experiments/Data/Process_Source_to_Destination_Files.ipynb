```python
import os
import pandas as pd
import numpy as np

# Define the path to the patient folder
patient_folder = "/content/Source"

# Define the path to the output folder
output_folder = "/content/Destination"

columns_needed = ['income', 'road_dist', 'cooking', 'y_am_pef', 'tempin', 'humidin', 'pm25in', 'co2in', 'tempdiffin', 'humidiffin', 'pm25diffin', 'pm10', 'pm25', 'o3', 'no2', 'co', 'so2', 'temp', 'windsd', 'humid', 'varp', 'dewpt', 'airp', 'seap', 'solrhr', 'solramnt', 'grdt', 'class']
```

```python
# Create an empty DataFrame to store the additional information
csv_lengths_df = pd.DataFrame(columns=["Patient ID", "CSV1 Length", "CSV2 Length", "Total Records"])

# Get a list of all patient folders inside the main patient folder
patient_folders = [folder for folder in os.listdir(patient_folder) if os.path.isdir(os.path.join(patient_folder, folder))]

# Iterate over each patient folder
for patient in patient_folders:

    combined_df = pd.DataFrame()
    pid_folder = os.path.join(patient_folder, patient)

    # Get a list of all CSV files in the PID folder
    csv_files = [file for file in os.listdir(pid_folder) if file.endswith(".csv")]

    # Create variables to store lengths
    csv1_length = 0
    csv2_length = 0

    # Iterate over each CSV file
    for csv_file in csv_files:
        csv_path = os.path.join(pid_folder, csv_file)

        df = pd.read_csv(csv_path)

        if "normalized_AM" in csv_file:
            csv1_length = len(df)

        if "normalized_PM" in csv_file:
            csv2_length = len(df)

        # Concatenate the current DataFrame with the combined DataFrame
        combined_df = pd.concat([combined_df, df])

    # Filter only necessary columns
    combined_df = combined_df.filter(columns_needed)

    # Shuffle the rows
    shuffled_df = combined_df.sample(frac=1, random_state=42).reset_index(drop=True)

    # Convert the combined DataFrame to a NumPy array
    combined_array = shuffled_df.to_numpy()

    # Calculate the total records
    total_records = len(shuffled_df)

    patient_output_folder = os.path.join(output_folder, patient)
    os.makedirs(patient_output_folder, exist_ok=True)

    # Define the filename for the output NPY file
    output_file = os.path.join(patient_output_folder, patient + ".npy")

    # Save the combined array as an NPY file
    np.save(output_file, combined_array)

    # Append the additional information to the csv_lengths_df DataFrame
    csv_lengths_df = csv_lengths_df.append({"Patient ID": patient,
                                            "CSV1 Length": csv1_length,
                                            "CSV2 Length": csv2_length,
                                            "Total Records": total_records}, ignore_index=True)

# Print the csv_lengths_df DataFrame
#print(csv_lengths_df)
```

```python
# sort df
csv_lengths_df = csv_lengths_df.sort_values(by="Patient ID")
```
